Exploratory Data Analysis (EDA) for Zomato's Kolkata Restaurants

Introduction:
Zomato has revolutionized the way people discover and experience food, offering a seamless platform to order food, check reviews, and gather comprehensive information about restaurants. In this analysis, we delve into the Kolkata restaurant landscape to derive actionable insights and provide strategic recommendations.

Data Overview:
The dataset comprises various attributes that describe the characteristics of restaurants, their services, and customer feedback. Here’s a detailed breakdown of the columns:

Name: The name of the restaurant.

URL: The restaurant’s URL on Zomato.

Cuisines: The variety of cuisines offered by the restaurant.

Area: The geographical location of the restaurant.

Timing: The restaurant’s operating hours.

Full_Address: The exact address of the restaurant.

PhoneNumber: The contact details of the restaurant.

IsHomeDelivery: Indicates the availability of home delivery (1 for available, 0 for not available).

isTakeaway: Indicates the availability of takeaway services (1 for available, 0 for not available).

isIndoorSeating: Indicates the availability of indoor seating (1 for available, 0 for not available).

isVegOnly: Indicates whether the restaurant is exclusively vegetarian (1 for vegetarian, 0 for not vegetarian).

Dinner Ratings: Customer ratings for dinner on a scale of 5.

Dinner Reviews: The number of customer reviews based on dinner experiences.

Delivery Ratings: Customer ratings for delivery services on a scale of 5.

Delivery Reviews: The number of customer reviews based on delivery experiences.

KnownFor: Signature delicacies that the restaurant is known for.

PopularDishes: High-rated dishes from the restaurant.

PeopleKnownFor: Unique aspects of the restaurant highlighted in reviews.

AverageCost: The average cost per visit.

Objective:
Our goal is to conduct a comprehensive Exploratory Data Analysis (EDA) to extract meaningful insights and offer strategic recommendations. By examining the dataset, we aim to understand the restaurant landscape, customer preferences, and service quality in Kolkata.

Data Cleaning

```python
#Removing these columns from the dataset
df.drop(['URL','Full_Address','PhoneNumber','KnownFor','Dinner Reviews','Delivery Reviews','Timing'],axis=1,inplace=True)
df.head()
```

```python
#Replacing the '-' to '0' from the Dinner Ratings column
df['Dinner Ratings'] = df['Dinner Ratings'].replace('-','0')
df['Dinner Ratings']
```

```python
#Replacing the '-' to '0' from the Delivery Ratings column
df['Delivery Ratings'] = df['Delivery Ratings'].replace('-','0')
df['Delivery Ratings']
```

```python
#Converting the datatype of Dinner Ratings column from object to float.
df['Dinner Ratings'] = df['Dinner Ratings'].astype('float64')
```

```python
#Filling the missing data present in the 'PopularDishes' column.
df['PopularDishes']=df['PopularDishes'].fillna('No Data')
```

```python
#Filling the missing data present in the 'PeopleKnownFor' column.
df['PeopleKnownFor']=df['PeopleKnownFor'].fillna('No Data')
```

```python
df.info()
```

We removed the columns URL, Full_Address, PhoneNumber, KnownFor, Dinner Reviews, Delivery Reviews, and Timing since they weren't needed for this analysis.

We replaced - with 0 in the Dinner Ratings and Delivery Ratings columns to change their datatype from object to float for numerical calculations.

We filled the missing data in the PopularDishes and PeopleKnownFor columns with 'No Data' to ensure there are no missing values and make analysis easier.

After these steps, our dataset is clean and ready for visualization, with no null values or messy data.

DATA VISUALISATION

1. Top 25 Areas having highest number of restaurants

```python
colors = sns.color_palette('Set2', 25)

plt.figure(figsize=(14, 10))
df['Area'].value_counts().head(25).plot(kind='bar', color=colors, edgecolor='black')

plt.title("Number of Restaurants by Area", fontsize=20, fontweight='bold')
plt.xlabel("Area", fontsize=16, fontweight='bold')
plt.ylabel("Number of Restaurants", fontsize=16, fontweight='bold')

plt.xticks(rotation=45, ha='right', fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.7)

for index, value in enumerate(df['Area'].value_counts().head(25)):
    plt.text(index, value + 3, str(value), ha='center', fontsize=12, fontweight='bold')

plt.show()
```

From this we conclude that the highest number of restaurants are present in New Town, Kolkata.

2. Pie-Chart representation of common cuisines available in most restaurants.

```python
colors = sns.color_palette('pastel')[0:10]

plt.figure(figsize=(10, 6))
df['Cuisines'].value_counts().head(10).plot(kind='pie', 
                                            colors=colors,
                                            autopct='%1.1f%%',
                                            startangle=140,
                                            explode=[0.1 if i == 0 else 0 for i in range(10)],
                                            shadow=True)

plt.title('Commonly Available Cuisines', fontsize=16, fontweight='bold')
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle

plt.legend(title='Cuisines', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='large')

plt.show()
```

From the above graph we can see ten cuisines available in most restaurants.

10. Top fifteen areas having highest ratings for deliveries.

```python
df['Delivery Ratings'] = pd.to_numeric(df['Delivery Ratings'], errors='coerce')

df = df.dropna(subset=['Delivery Ratings'])

top15_area = df.groupby('Area')['Delivery Ratings'].mean().sort_values(ascending=False).head(15).reset_index()

plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")

sns.barplot(x='Area', y='Delivery Ratings', data=top15_area, palette='viridis', hue='Area', dodge=False, legend=False)

plt.title("Area-wise Delivery Ratings", fontsize=20, fontweight='bold')
plt.xlabel("Areas", fontsize=16, fontweight='bold')
plt.ylabel("Delivery Ratings", fontsize=16, fontweight='bold')
plt.xticks(rotation=90, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Answering Few Questions.

1. What are the top Ten restaurants for each category (Dine-In, Delivery, Takeaway, Vegetarian)?

```python
# Top ten restaurants for Dine-in category
top_10_dine_in = df.loc[(df['isIndoorSeating'] == 1) & (df['Dinner Ratings'] >= 4.5) & (df['Delivery Ratings'] >= 4)].sort_values(by=['Dinner Ratings', 'Delivery Ratings'], ascending=False)[:10]

top_10_dine_in
```

```python
# Top ten restaurants for Delivery category
top_10_restaurants = df.loc[(df['IsHomeDelivery']==1) & (df['Dinner Ratings'] >= 4.5)
& (df['Delivery Ratings'] >= 4)].sort_values(by=['Dinner Ratings', 'Delivery Ratings'], ascending=False)[:10]

top_10_restaurants
```

```python
# Top ten restaurants for Takeaway category
top_10_takeaway = df.loc[(df['isTakeaway']==1) & (df['Dinner Ratings'] >= 4.5)
& (df['Delivery Ratings'] >= 4)].sort_values(by=['Dinner Ratings', 'Delivery Ratings'], ascending=False)[:10]

top_10_takeaway
```

```python
# Top ten restaurants for Vegetarian category
top_10_vegetarian = df.loc[(df['isVegOnly']==1) & (df['Dinner Ratings'] >= 4.5)
& (df['Delivery Ratings'] >= 4)].sort_values(by=['Dinner Ratings', 'Delivery Ratings'], ascending=False).head(10)

top_10_vegetarian
```

2. What are the highest rating restaurants for Chinese cuisine?

```python
df.loc[(df['Cuisines']=='Chinese') & (df['Dinner Ratings'] >= 4)
& (df['Delivery Ratings'] >= 4)].sort_values(by=['Dinner Ratings','Delivery Ratings'], ascending=False)[:5]
```

3. Which cuisines have highest rating in different localities?

```python
cuisine_ratings = df.groupby(['Area', 'Cuisines'])[['Dinner Ratings','Delivery Ratings']].mean().reset_index()

top_12_area = cuisine_ratings.sort_values(by='Dinner Ratings', ascending=False).head(12)

top_12_area
```

4. Which restaurant have highest rating in different localities in Kolkata?

```python
rest_rating = df.groupby(['Area', 'Name'])[['Dinner Ratings','Delivery Ratings']].mean().reset_index()

top_12_rest = rest_rating.sort_values(by=['Dinner Ratings'], ascending=False).head(12)

top_12_rest
```
